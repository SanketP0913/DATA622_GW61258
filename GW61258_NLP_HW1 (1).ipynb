```python
import requests
from bs4 import BeautifulSoup
from bs4.element import Comment

def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True

def text_from_html(html):
    soup = BeautifulSoup(html, 'html.parser')
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)
    return u" ".join(t.strip() for t in visible_texts if t.strip())

# 1. Visit Maryland site
md_url = 'https://www.visitmaryland.org/'
md_resp = requests.get(md_url)
print('Status code:', md_resp.status_code)

# 2. Visible text
visible_text = text_from_html(md_resp.text)
print(' Visible text from visitmaryland.org ')
print(visible_text[:500] + '...')  # print first 500 characters for brevity

# 3, 4, 5. Wikipedia NLP page
wiki_url = 'https://aws.amazon.com/what-is/nlp/?trk=80384a35-d261-4dc4-ae07-058697ea891c&sc_channel=ps&ef_id=CjwKCAjw2vTFBhAuEiwAFaScwsltxpnEVmjqsbC5Bju1mV1yK8jZnuDL_kK6ybU3rtN4oLAmALT-kBoCsXkQAvD_BwE:G:s&s_kwcid=AL!4422!3!751258939134!e!!g!!natural%20language%20processing!22542047959!178682824305&gad_campaignid=22542047959&gbraid=0AAAAADjHtp_MFhUn3o4MiagYCLoJ3DbAn&gclid=CjwKCAjw2vTFBhAuEiwAFaScwsltxpnEVmjqsbC5Bju1mV1yK8jZnuDL_kK6ybU3rtN4oLAmALT-kBoCsXkQAvD_BwE'
wiki_resp = requests.get(wiki_url)
soup = BeautifulSoup(wiki_resp.text, 'html.parser')

print('\n Headings on NLP page ')
for tag in ['h1', 'h2', 'h3']:
    for h in soup.find_all(tag):
        print(f'{tag}: {h.get_text().strip()}')

print('\n Links on NLP page ')
for a in soup.find_all('a', href=True):
    print(a['href'])

#intro
first_p = soup.find('p')
intro = first_p.get_text().strip() if first_p else ''
with open('nlp_intro.txt', 'w', encoding='utf-8') as f:
    f.write(intro)
print('\nFirst paragraph saved to nlp_intro.txt')
```
